```python
breakfast_stable = ["гранола протеиновая", "авокадо тост", "овсянные панкейки", "овсянные шоколадные панкейки",
             "панкейки", "ролл яйцо и ветчина", "чиа пудинг", "тост с тунцом", "конвертики с сыром и зеленью"]
how_to_eggs = ["омлет", "яичница", "скрембл (болтунья)", "пашот", "яйцо всмятку", "яйцо вкрутую"]
eggs_carbons = ["Ц/з хлеб", "хлебцы", "лаваш", "кукуруза", "хумус"]
eggs_cellulose = ["помидор", "перец", "огурец", "спаржа", "брокколи", "авокадо (1/2 шт)"]

salad_base = ["морковь", "свекла", "кукуруза", "горох", "нут", "фасоль", 
         "чечевица", "кус-кус", "булгур", "бурый рис", "киноа", "паста"]
salad_protein = ["курица", "морепродукты", "тунец", "форель", "орехи", "сыр фета", "моцарелла"]
salad_cellulose = ["листья салата", "огурец", "помидор", "сладкий перец", "брокколи",  "спаржа", "авокадо"]


lunch_protein = ["курица", "индейка", "морепродукты", "рыба белая", "рыба красная", "грибы"]
lunch_carbs = ["ц/з хлеб", "картофель", "свекла", "морковь", "брокколи", "кукуруза", "бобовые", 
               "гречка", "киноа", "бурый рис", "булгур", "паста", "лапша"]
lunch_cellulose = ["помидор", "перец", "огурец", "спаржа", "брокколи", "листья салата", "авокадо"]

snack_stable = ["орехи", "изюм","овощи и хумус", "фрукты", "финики"]
```

```python
for i in range(8):
    breakfast = random.choice(breakfast_stable + 
                              list(itertools.product(how_to_eggs, 
                                                     eggs_carbons, 
                                                     eggs_cellulose)))
    salad = random.choice(list(itertools.product(salad_base, salad_protein, salad_cellulose)))
    lunch = random.choice(list(itertools.product(lunch_protein, lunch_carbs, lunch_cellulose)))
    snack = random.choice(snack_stable)
    print(f"""День {i+1}
    Завтрак: {breakfast}
    Обед: {lunch}
    Ужин: {salad}
    Перекус: {snack}
    """)
```
